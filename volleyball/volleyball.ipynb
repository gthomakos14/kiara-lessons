```python
def process_link(link):
    result = requests.get(link)
    # assert result.status_code == 200, f'Status code was {result.status_code}'
    if result.status_code != 200:
        print(f'Link {link} has status code {result.status_code}')
        return None
    json = result.json()
    if json['meta']['status'] == 'In Progress':
        return None
    team_ids = [str(i['id']) for i in json['meta']['teams']]
    team_names = [i['shortName'] + ' ' + i['nickName'] for i in json['meta']['teams']]
    team_dict = {k: v for (k, v) in zip(team_ids, team_names)}
    
    teams = json['teams']
    frames_list = []
    for i in json['teams']:
        df = pd.DataFrame(i['playerStats'])
        df['teamId'] = str(i['teamId'])
        df['teamName'] = df['teamId'].replace(team_dict)
        frames_list.append(df)
    df_stats = pd.concat(frames_list)
    
    return df_stats
```

```python
processed_links = []
for i in links:
    processed_links.append(process_link(i))
df_final = pd.concat(processed_links)
```

```python
df_final.to_csv('kiara_volleyball.csv', index=False)
```
